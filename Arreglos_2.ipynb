```python
# diagonal superio , e inferior
import numpy as np
tamano=4
# crearemos una matriz de 4*4 para llenarlos de ceros
matriz=np.zeros((tamano,tamano))
print(f"[#f70071]{matriz}[/#f70071]")
# diagonal superior
#i= fila
#j = columna
for f in range(tamano):
  for c in range(tamano):
    if f<=c:
      matriz[f,c]=1
print(f"[#ff1b82]{matriz}[/#ff1b82]")
# funcion de numpy con diagonal
matriz=np.triu(matriz)
print(f"[#ff5aa1]{matriz}[/#ff5aa1]")
#diagonal inferior
matriz=np.tril(matriz)
print(f"[#ff96c5]{matriz}[/#ff96c5]")
for f in range(tamano):
  for c in range(tamano):
    if c<=f:
      matriz[f,c]=1
print(f"[#ffc0dc]{matriz}[/#ffc0dc]")
```
